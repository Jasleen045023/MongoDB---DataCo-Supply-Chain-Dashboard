# OBJECTIVES

The dashboard has been created on MongoDB Atlas, and the database was sourced from Kaggle and subsequently imported into MongoDB Compass for analysis and visualization. With its rich and varied features, the DataCo dataset is particularly suited for real-world scenarios requiring data-driven decision-making, such as campaign planning, regional performance evaluation, and product-level analysis.

The dashboard has been preapred on the basis of the following objectives, to:
1. Understand the geographical distribution of sales to identify high-performing regions and areas with untapped potential.
2. Analyze sales performance by region to prioritize marketing investments.
3. Assess the relationship between delivery performance and sales.
4. Identify shipping methods that pose a risk of late delivery to improve customer satisfaction.
5. Identify key regions and order statuses contributing to profits for targeted campaigns.
6. Visualize how sales volumes vary across different order statuses to optimize processes.
7. Compare department-wise performance to allocate marketing budgets more effectively.

```python
# Identify the top-selling products
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print("Top-Selling Products:\n", top_products)
```

```python
# Identify repeat customers
repeat_customers = df['Customer Id'].value_counts().loc[lambda x: x > 1].count()
print(f"Number of Repeat Customers: {repeat_customers}")

# Calculate the percentage of repeat customers
total_customers = df['Customer Id'].nunique()
repeat_customer_percentage = (repeat_customers / total_customers) * 100
print(f"Percentage of Repeat Customers: {repeat_customer_percentage:.2f}%")
```

```python
# Analyze average shipping times
shipping_efficiency = df['Days for shipping (real)'] - df['Days for shipment (scheduled)']
print("Average Shipping Delay (Days):", shipping_efficiency.mean())
```

```python
# Calculate the percentage of completed orders
completed_orders = df[df['Order Status'] == 'Complete'].shape[0]
total_orders = df.shape[0]
completion_rate = (completed_orders / total_orders) * 100
print(f"Order Completion Rate: {completion_rate:.2f}%")
```
